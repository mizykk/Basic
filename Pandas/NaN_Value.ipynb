# 결측값 다루기

```python
import pandas as pd
import seaborn as sns
```

예시데이터 불러오기

```python
pg = sns.load_dataset('penguins')
pg.head()
```

### 결측값 확인하기
- data.isna() : 결측값 True
- data.isnull() : 결측값 True
- data.notna() : 결측값 False
- data.notnull() : 결측값 False

```python
# 결측값 True
pg.isna()
```

```python
# 결측값 True
pg.isnull()
```

```python
# 결측값 False
pg.notna()
```

```python
# 결측값 False
pg.notnull()
```

### 몇 개 있는지 알고 싶다.

결측값 기준

```python
pg.isna().sum()
```

```python
pg.isnull().sum()
```

결측값이 아닌 값 기준

```python
pg.notnull().sum()
```

```python
pg.notna().sum()
```

### 결측값이 들어있는 행만

```python
pg[pg.isnull().any(axis = 1)]
```

### 결측값이 없는 행만

```python
pg[pg.notnull().all(axis = 1)]
```

### 결측값이 있는 행 제거

```python
pg.dropna(axis = 0)
```

### 결측값 채우기

```python
pg.fillna('-')
```
