# [Seaborn](https://seaborn.pydata.org/index.html)   

```
import seaborn as sns
```

```
# 예시데이터
tips = sns.load_dataset("tips")
tips.head()
```

### [**sns.lineplot(x = , y = , hue = , data = )**](https://seaborn.pydata.org/generated/seaborn.lineplot.html) : 선 그래프  

```
sns.lineplot(x = 'size', y = 'tip', hue = 'time', data = tips, markers = True)
```

### [sns.regplot(x = , y = , data = )](https://seaborn.pydata.org/generated/seaborn.regplot.html?highlight=regplot) : 산점도  

```
sns.regplot(x = 'total_bill', y = 'tip', data = tips)
```

### [sns.boxplot(x = , y = , hue = , data = )](https://seaborn.pydata.org/generated/seaborn.boxplot.html?highlight=boxplot) : 상자 그래프  

```
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips, palette="Set3")
```

### [**sns.pairplot(data = , hue = )**](https://seaborn.pydata.org/generated/seaborn.pairplot.html) : 산점도 행렬

```
sns.pairplot(data = tips)
```

### [sns.heatmap](https://seaborn.pydata.org/generated/seaborn.heatmap.html) : 히트맵

```
flights = sns.load_dataset("flights")
flights = flights.pivot("month", "year", "passengers")
ax = sns.heatmap(flights, annot=True, fmt="d")
```

### [sns.distplot(x, bins = )](https://seaborn.pydata.org/generated/seaborn.distplot.html)

*seaborn.distplot(a, bins=None, hist=True, kde=True, rug=False, fit=None, hist_kws=None, kde_kws=None, rug_kws=None, fit_kws=None, color=None, vertical=False, norm_hist=False, axlabel=None, label=None, ax=None)*

```
sns.distplot(tips['total_bill'])
```

```
sns.distplot(tips['total_bill'], bins = 50)
```

```
from scipy.stats import norm
sns.distplot(flights['passengers'], fit = norm)
```
